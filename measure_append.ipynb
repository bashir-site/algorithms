```python
import matplotlib.pyplot as plt
import time
```

```python
times = []
```

```python
list1 = []
while len(list1) < 100:
    start_time = time.time()
    list1.append(1)
    end_time = time.time()
    times.append(end_time - start_time)
```

```python
plt.bar(range(len(times)), times)
plt.xlabel('Итерации')
plt.ylabel('Время')
plt.title('Затраченное время на операцию APPEND')
plt.show()
```
